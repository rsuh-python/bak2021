Все эти задачи можно решать с помощью любого понравившегося вам сайта для проверки РВ; мне достаточно прислать сами РВ в любом виде (только будьте осторожны, телеграм может позаменять некоторые символы внутри регулярок, поэтому лучше в каком-либо файлике, хоть картинкой, хоть текстом)

### Задача 1. 

Найдите в тексте все вхождения слова Python в любом регистре (и заборчиком тоже). Ответьте на вопрос: нужна ли для этого вообще регулярка?

### Задача 2.

Найдите в тексте все вхождения слов "мама" и "папа" с помощью одного регулярного выражения.

### Задача 3.

Найдите в тексте все словоформы для леммы "мама" (мамой, маме, мамы...)

### Задача 4.

Найдите в тексте все арабские числа (цифр может быть сколько угодно)

### Задача 5.

Найдите в тексте все смайлики вида :), :(, :-), :-(.

### Задача 6.

Найдите в тексте все википедийные сноски вида [1].

### Задача 7.

Найдите в тексте все словоформы для леммы "точка", но таким образом, чтобы "цветочек" и "деточка" нас не устраивали.

### Задача 8.

Найдите в тексте все вариации мема "ну ... и ..." (то есть, вместо ... любое слово: ну питон и питон, ну регулярка и регулярка)

### Задача 9.

Найдите в тексте все содержимое скобок. Вложенных скобок в нашем тексте, считаем, нет. Внутри скобок может быть вообще все, что угодно.

### Задача 10.

Найдите в тексте все слова, которые начинаются с заглавной буквы и состоят при этом из строчных, но не находятся в начале строки.

### Задача 11.

Найти в тексте все смайлики вида ^_^, *_*, T_T. Осторожнее: смайлики вида ^_* нам не нужны!

### Задача 12.

Составьте шаблон, который будет захватывать все кириллические слова. Мы считаем, что слово может состоять только из русских букв, включая ё, и иметь внутри дефисы. Шаблон не должен захватывать дефис на краю слова, а также слова, внутри которых два дефиса подряд.
