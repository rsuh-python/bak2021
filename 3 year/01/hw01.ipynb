### Задание 1.

Напишите скрипт в формате .py, в котором определите три функции:

1. Функция, которая проверяет, является ли строка кириллическим словом (можно использовать старый код); возвращает True или False
2. Функция, которая считывает текст из файла по пути path, делит его на слова и очищает от знаков препинания; возвращает список слов
3. Функция, которая принимает список слов и считает его лексическое разнообразие в процентах

Импортируйте скрипт в новый файл, в котором нужно обработать любой текстовый файл и получить количество кириллических слов (таких, которые содержат только кириллические буквы и не содержат ни дефисов, ни других символов) и лексическое разнообразие. 

### Задание 2 (2 балла).

Напишите игру. Когда пользователь выбирает действие "новый раунд", игра предлагает ему ввести имена для бойцов, потом между двумя бойцами происходит бой, который заканчивается, если у одного из них закончилось здоровье. Игра работает в цикле (можно играть бесконечное количество раундов) и выходит, когда пользователь выбирает действие "выход". 

Логика самой игры такая: 

- Должен быть класс Fighter, у которого есть атрибуты health, armor, strength и метод экземпляра класса hit(self, other). Может еще быть атрибут name.
- Проследите, чтобы не вышло так, что у одного из бойцов броня (armor) выше, чем сила удара второго! Это будет бесконечная игра. 
- Можно задавать все три атрибута с помощью функции randrange() или randint() из модуля random.
- В методе hit у бойца, которого ударили, здоровье вычисляется по формуле: -= (сила бьющего бойца - броня ударенного бойца)
- Сам бой функционирует в цикле, который выполняется до тех пор, пока у обоих бойцов есть здоровье. 
- Заведите переменную, которая будет принимать случайным образом значения 0 и 1 (или 1 и 2) - random вам поможет. В зависимости от этой переменной будет бить то один, то другой боец. Это будет наш игральный кубик.
