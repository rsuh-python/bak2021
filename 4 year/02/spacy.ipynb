```python
import spacy 
from spacy import displacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = '''ASCII was developed from telegraph code.'''
doc = nlp(text)
```

```python
for token in doc:
  print(token.text, token.lemma_, token.pos_, token.morph, token.dep_, token.head.text)
```

```python
for ent in doc.ents:
  print(ent.text, ent.label_)
```

```python
displacy.render(doc, style='dep', jupyter=True)
```

```python
displacy.render(doc, style="ent", jupyter=True)
```
